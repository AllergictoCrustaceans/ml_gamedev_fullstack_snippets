## matplotlib example workflow

```python
# 0. import matplotlib, get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save figure and show
fig.savefig('./sample-plot.png')
```

### Create Subplots

```python
# method 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
## plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# method 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Which Graphing API to Use? (Pyplot V. Matplotlib OO method?)

- If you want a quick viz, use pyplot method
- If you want to plot something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method -- plotting directly on Pandas DF
over_50.plot(kind="scatter", x = "age", y = "chol", c='target');
```

```python
# OO Method w/ pyploy method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x='age', y="chol", c='target', ax=ax);

ax.set_xlim([45, 100]); # set the x- range of your axes
```

```python
# Set style
plt.style.use("seaborn-v0_8-whitegrid")

## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target'],
                    cmap="winter"); # cmap changes the color scheme

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add legend
ax.legend(*scatter.legend_elements(), title="Target"); # * means unpack the elements

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

## Subplot of Chol, Age, Thalach

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], 
                      y=over_50['chol'], 
                      c=over_50['target'],
                      cmap="winter");

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Change x axis range
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle="--");

# Add data to ax1
scatter1 = ax1.scatter(x=over_50['age'], 
                       y=over_50['thalach'], 
                       c=over_50['target'], 
                       cmap="winter");

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Change ax1 x axis range
ax1.set_xlim([50, 80])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle="--");

# Add a main title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
