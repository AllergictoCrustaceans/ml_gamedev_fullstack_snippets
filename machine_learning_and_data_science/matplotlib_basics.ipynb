```python
# 1st method:
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method:
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1]) # add axes
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd method: (Recommended)
fig, ax = plt.subplots()
ax.plot(x, y); # add some data
type(fig), type(ax)
```

## matplotlib example workflow

```python
# 0. import matplotlib, get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save figure and show
fig.savefig('./sample-plot.png')
```

## Making Figures with Numpy Arrays

- line plot
- scatter plot
- bar plot
- histogram
- sub plot

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data, create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# use same data, create scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# create a plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8, 
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# create histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Create Subplots

```python
# method 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
## plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# method 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

## Plotting Data from Pandas

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales= pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.rand(1000),
              index=pd.date_range('1/1/2023', periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales["Price"]
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# plot total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```

```python
# Reassign price column to int
car_sales["Price"] = car_sales["Price"].astype(int)

# plot scatter
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
